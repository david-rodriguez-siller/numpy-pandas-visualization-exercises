Curves to graph:

- y = 0n + 1 and label the curve 'O(1)'
- y = log(n) and label the curve 'O(log n)'
- y = n and label the curve 'O(n)'
- y = n * log(n) and label it 'O(n log n)'
- y = $n^2$ and label it 'O($n^2$)'
- y = $2^n$ and label it 'O($2^n$)'
- y = n! and label it 'O(n!)'
- y = $n^n$ and label it 'O($n^n$)'

```python
#import pyplot module as alias plt
import matplotlib.pyplot as plt  

#import other libraries 
import numpy as np
import math
from random import randint
import random
import pandas as pd
```

```python
plt.figure(figsize = (16, 12))

x = range(0, 25)
y1 = [0 * n + 1 for n in x]
y2 = [np.log(n) for n in x]
y3 = [n for n in x]
y4 = [n * np.log(n) for n in x]
y5 = [n ** 2 for n in x]
y6 = [2 ** n for n in x]
y7 = [math.factorial(n) for n in x]
y8 = [n ** n for n in x]

plt.plot(n, y1, label = 'O(1)', color = 'crimson')
plt.plot(n, y2, label = 'O(log n)', color = 'steelblue')
plt.plot(n, y3, label = 'O(n)', color = 'darkgoldenrod')
plt.plot(n, y4, label = 'O(n log n)', color = 'olive')
plt.plot(n, y5, label = 'O($n^2$)', color = 'teal')
plt.plot(n, y6, label = 'O($2^n$)', color = 'blueviolet')
plt.plot(n, y7, label = 'O(n!)', color = 'cadetblue')
plt.plot(n, y8, label = 'O($n^n$)', color = 'firebrick')

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.grid('True', ls = '--')
plt.xlim(0, 25)
plt.ylim(0, 25)

plt.legend();
```
