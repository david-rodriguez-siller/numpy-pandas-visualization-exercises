4. Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
def get_db_url(db_name):
    from env import user, host, password
    return f'mysql+pymysql://{user}:{password}@{host}/{db_name}'
```

```python
url = get_db_url('chipotle')
sql = 'SELECT * FROM orders;'
chipotle = pd.read_sql(sql, url)
chipotle.head()
```

```python
chipotle['item_price'] = chipotle.item_price.str.replace('$', '').astype('float')
```

```python
best_sellers = chipotle.groupby('item_name').quantity.sum().nlargest(4)
best_sellers
```

```python
chipotle.groupby('item_name').quantity.sum().sort_values(ascending = False).head(4)
```

```python
revenue = chipotle[chipotle.item_name.isin(best_sellers.index)].groupby('item_name').item_price.sum()
revenue = revenue.sort_values(ascending = False).reset_index()
revenue
```

```python
revenue = chipotle[chipotle.item_name.isin(best_sellers.index)].groupby('item_name').item_price.sum()
revenue = revenue.sort_values(ascending = False).reset_index()
revenue
```

```python
plt.figure(figsize = (8, 6))
sns.barplot(data = revenue, y = 'item_name', x = 'item_price')
plt.title('Most Popular Items & Revenue')

plt.ylabel('')
plt.xlabel('Gross Revenue')
plt.suptitle('Revenue From Top Four Selling Items');
```
