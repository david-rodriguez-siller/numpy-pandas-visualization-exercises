#### 1. Use matplotlib to plot the following equation: y = x^2 - x + 2
#### You'll need to write the code that generages the x and y points.
#### Add an anotation for the point 0, 0 region

```python
#import pyplot module as alias plt
import matplotlib.pyplot as plt  

#import other libraries 
import numpy as np
import math
from random import randint
import random
import pandas as pd
```

```python
x = range(-10, 11)
y = [n ** 2 - n + 2 for n in x]
plt.plot(x, y)

plt.title('$y = x^2 - x + 2$')
plt.xlabel('$x$');
plt.ylabel('$x^2 - x + 2$');

plt.annotate('This is is 0,0', (0, 0), (4, 88), arrowprops = dict());
```

#### 2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
1. y = $\sqrt{x}$
2. y = $x ^ 3$
3. y = $2 ^ x$
4. y = $1/(x + 1)$

#### You can use functions from the math module to help implement some of the equations above.

 - Graph #1: $y = \sqrt{x}$

```python
x = range(0, 50)
y = [math.sqrt(n) for n in x]
plt.plot(x,y);

plt.title('$y = \sqrt{x}$')
plt.xlabel('$x$');
plt.ylabel('$\sqrt{x}$');
```

 - Graph #2: $y = x^3$

```python
x = range(-50, 50)
y = [n * n * n for n in x]
plt.plot(x,y);

plt.title('$y = x^3$')
plt.xlabel('$x$');
plt.ylabel('$x^3$');
```

 - Graph #3: $y = 2^x$

```python
x = range(-10, 10)
y = [2 ** n for n in x]
plt.plot(x,y);

plt.title('$y = 2^x$')
plt.xlabel('$x$');
plt.ylabel('$2^x$');
```

 - Graph #4: $1/(x + 1)$

```python
x = range(0, 30)
y = [1/(n + 1) for n in x]
plt.plot(x,y);

plt.title('$1/(x + 1)$')
plt.xlabel('$x$');
plt.ylabel('$1/(x+1)$');
```

#### 3. Combine the figures you created in the last step into one large figure with 4 subplots

```python
x = range(0, 50)
y1 = [math.sqrt(n) for n in x]
y2 = [n * n * n for n in x]
y3 = [2 ** n for n in x]
y4 = [1/(n + 1) for n in x]

plt.figure(figsize = (8,6))

plt.suptitle('Exercise Math Functions', fontsize = 14)

plt.subplot(2,2,1)
plt.plot(x,y1)
plt.text(0.5, 0.5, str((1)), ha = 'center', fontsize = 14)

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.text(0.5, 0.5, str((2)), ha = 'center', fontsize = 14)

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.text(0.5, 0.5, str((3)), ha = 'center', fontsize = 14)

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.text(0.5, 0.5, str((4)), ha = 'center', fontsize = 14);
```

#### 4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
x = range(0, 15)
y1 = [math.sqrt(n) for n in x]
y2 = [n * n * n for n in x]
y3 = [2 ** n for n in x]
y4 = [1/(n + 1) for n in x]

plt.figure(figsize = (8,6))

plt.suptitle('Exercise Math Functions', fontsize = 14)

plt.plot(x,y1, color = 'orange', marker = '*', linestyle = 'dotted', label = '$\sqrt{x}$')
plt.plot(x,y2, color = 'purple', marker = '.', linestyle = 'dashdot', label = '$x ^ 3$')
plt.plot(x,y3, color = 'teal', marker = 'o', linestyle = '-', label = '$2 ^ x$')
plt.plot(x,y4, color = 'blue', marker = '', linestyle = '', label = '$1/(x+1)$')

# add x and y axis labels
plt.xlabel('x')
plt.ylabel('f(x)')

plt.grid('True', ls = '--')

plt.legend();
```
